# Analysis of the Gaussian Probability Density Function

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

### Random sampling from a normal distribution

```python
# Define distribution parameters 
numSamples = 500
mu = 0
sigma = 0.1

# Draw random samples from a normal distribution
samples = np.random.normal(mu,sigma,numSamples)

# Plot distribution 
count, bins, ignored = plt.hist(samples, 30, density=True, color='darkblue')
plt.title('Samples')
x = bins[:-1]
data = count
```

## Unmodified Gaussian Function 
<li> Symmetric around x = mu
<li> Area bound under the curve always = 1
<li> Points of inflection @ x-sigma and x+sigma 

```python
def fit_gauss(x, data, par, flag):
    # free parameters (mu and sigma)
    mu = par[0]
    sigma = par[1]
    
    # objective function 
    fit = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (x - mu)**2 / (2 * sigma**2))
    
    # cost function 
    cost = sum(data-fit)**2
    if flag == True:
        return [fit, cost]
    else:
        return cost
```

```python
plt.figure()
plt.plot(x, data, 'ko')

y, sse = fit_gauss(x, data, [mu, sigma], True)
plt.plot(x, y, color='gray')
plt.title('Unmodified Gaussian Probability Density Function')
```

## Removing 1/sigma factor 
<li> Controls height/amplitude of the curve 
<li> Factor acts to increase the gain since 1/sigma > 1 

```python
def fit_gauss2(x, data, par, flag):
    # free parameters (mu and sigma)
    mu = par[0]
    sigma = par[1]
    
    # objective function 
    fit = 1/(np.sqrt(2 * np.pi)) * np.exp( - (x - mu)**2 / (2 * sigma**2))
    
    # cost function 
    cost = sum(data-fit)**2
    if flag == True:
        return [fit, cost]
    else:
        return cost
```

```python
plt.figure()
plt.plot(x, data, 'ko')

y, sse = fit_gauss2(x, data, [mu, sigma], True)
plt.plot(x, y, color='red', label='Modified Function')
plt.title('Gaussian Probability Density Function Without 1/sigma factor')
y, sse = fit_gauss(x, data, [mu, sigma], True)
plt.plot(x, y, color='gray', label='Unmodifed Function')
plt.legend()
```

## Removing 1/sqrt(2pi) factor 
<li> Controls height/amplitude of curve 
<li> Factor acts to decrease the gain since 1/sqrt(2pi) < 1

```python
def fit_gauss3(x, data, par, flag):
    # free parameters (mu and sigma)
    mu = par[0]
    sigma = par[1]
    
    # objective function 
    fit = 1/(sigma) * np.exp( - (x - mu)**2 / (2 * sigma**2))
    
    # cost function 
    cost = sum(data-fit)**2
    if flag == True:
        return [fit, cost]
    else:
        return cost
```

```python
plt.figure()
plt.plot(x, data, 'ko')

y, sse = fit_gauss3(x, data, [mu, sigma], True)
plt.plot(x, y, color='red', label='Modified Function')
plt.title('Gaussian Probability Density Function Without 1/sqrt(2pi) factor')
y, sse = fit_gauss(x, data, [mu, sigma], True)
plt.plot(x, y, color='gray', label='Unmodified Function')
plt.legend()
```

## Removing square from exponent 
<li> Gives symmetry to curve 

```python
def fit_gauss5(x, data, par, flag):
    # free parameters (mu and sigma)
    mu = par[0]
    sigma = par[1]
    
    # objective function 
    fit = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (x - mu) / (2 * sigma))
    
    # cost function 
    cost = sum(data-fit)**2
    if flag == True:
        return [fit, cost]
    else:
        return cost
```

```python
plt.figure()
plt.plot(x, data, 'ko')

y, sse = fit_gauss5(x, data, [mu, sigma], True)
plt.plot(x, y, color='red', label='Modified Function')
plt.title('Gaussian Probability Density Function Without Square Factor in Exponent')
y, sse = fit_gauss(x, data, [mu, sigma], True)
plt.plot(x, y, color='gray', label='Unmodified Function')
plt.legend()
```

## Removing 1/Sigma^2 Factor from Exponent

## Removing -1/2 from Exponent 
<li> Flips curve upside-down

```python
def fit_gauss7(x, data, par, flag):
    # free parameters (mu and sigma)
    mu = par[0]
    sigma = par[1]
    
    # objective function 
    fit = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp((x - mu)**2 / (sigma**2))
    
    # cost function 
    cost = sum(data-fit)**2
    if flag == True:
        return [fit, cost]
    else:
        return cost
```

```python
plt.figure()
plt.plot(x, data, 'ko')

y, sse = fit_gauss7(x, data, [mu, sigma], True)
plt.plot(x, y, color='red', label='Modified Function')
plt.title('Gaussian Probability Density Function Without -1/2 Factor in Exponent')
y, sse = fit_gauss(x, data, [mu, sigma], True)
plt.plot(x, y, color='gray', label='Unmodified Function')
plt.legend()
```
